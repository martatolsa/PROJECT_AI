## Life Expectancy & Healthy Life Expectancy (HALE)
**La esperanza de vida** proporciona una estimaci√≥n de cu√°ntos a√±os, en promedio, se espera que viva una poblaci√≥n. Sin embargo, la **esperanza de vida saludable (HALE)** es un indicador m√°s preciso del verdadero estado de salud de una poblaci√≥n.

No se trata solo de la cantidad de a√±os que vive una persona en promedio, sino de cu√°ntos de esos a√±os puede esperar vivir con buena salud, sin enfermedades o discapacidades graves que afecten su calidad de vida.

Este indicador es crucial porque muestra la diferencia entre simplemente vivir m√°s a√±os y vivirlos en condiciones que permitan una vida plena y activa. Mientras que la esperanza de vida general mide la longevidad, **HALE tiene en cuenta factores como la carga de enfermedades, la discapacidad y el acceso a servicios de salud de calidad, proporcionando una visi√≥n m√°s completa del bienestar de una sociedad.**

## Tasa de mortalidad infantil y materna

Las tasas de mortalidad infantil y materna son indicadores clave del desarrollo social y del desempe√±o de los sistemas de salud. Reflejan no solo el acceso a servicios m√©dicos de calidad, sino tambi√©n las condiciones socioecon√≥micas y educativas de una poblaci√≥n.

La tasa de mortalidad infantil ‚Äîen particular la mortalidad en menores de cinco a√±os‚Äî mide el riesgo de que un ni√±o fallezca antes de cumplir cinco a√±os. En las √∫ltimas d√©cadas, se ha logrado un progreso significativo en la reducci√≥n de esta tasa, en gran parte gracias al aumento de la cobertura de vacunaci√≥n y la mejora en la atenci√≥n m√©dica perinatal. Sin embargo, los datos muestran que persisten grandes desigualdades: √Åfrica subsahariana sigue siendo la regi√≥n m√°s afectada, con una tasa m√°s de ocho veces superior a la de Europa.

## 3.1. Construcci√≥n de funciones para limpiar y preprocesar

Funci√≥n "preprocess_indicator" preprocesa un dataframe para un indicador espec√≠fico:
- Filtra seg√∫n el indicador
- Renombra las columnas
- Elimina columnas irrelevantes
- Filtra la columna sex por "Both sexes"
- Asegura los tipos de datos de cada columna
- Agrupa los datos y calcula la media para las combinaciones duplicadas

```python

def preprocess_indicator(df, indicator_name):
    """
    Preprocesa un dataframe para un indicador espec√≠fico:
    - Filtra seg√∫n el indicador
    - Renombra las columnas
    - Elimina columnas irrelevantes
    - Filtra la columna sex por "Both sexes"
    - Asegura los tipos de datos de cada columna
    - Agrupa los datos y calcula la media para las combinaciones duplicadas
    """
    # Filtrar datos seg√∫n el indicador
    df_filtered = df[df['Indicator'] == indicator_name]

    # Asegurarse de que 'Dim1' est√© en las columnas, si no crearla
    if 'Dim1' not in df_filtered.columns:
        df_filtered['Dim1'] = 'Both sexes'

    columns_to_keep = [
    "Indicator", "ParentLocationCode", "ParentLocation", "SpatialDimValueCode",
    "Location", "Period", "Dim1" ,"FactValueNumeric", "Value"
    ]
    df_filtered = df_filtered[columns_to_keep]

    # Renombrar las columnas
    df_filtered = df_filtered.rename(columns={
        'Indicator': 'indicator',
        'ParentLocationCode': 'continent_code',
        'ParentLocation': 'continent',
        'SpatialDimValueCode': 'country_code',
        'Location': 'country',
        'Period': 'year',
        'Dim1': 'sex',
        'FactValueNumeric': 'value',
        'Value': 'value_interval'
    })

    # Filtrar por "Both sexes"
    df_filtered = df_filtered[df_filtered['sex'] == 'Both sexes']
    # Filtrar desde 1990 en adelante
    df_filtered = df_filtered[df_filtered['year'] >= 1990]

    # Eliminar columnas irrelevantes
    df_filtered = df_filtered[['continent','country', 'year', 'value']]

    # Asegurar tipos
    df_filtered['continent'] = df_filtered['continent'].astype(str)
    df_filtered['country'] = df_filtered['country'].astype(str)
    df_filtered['year'] = df_filtered['year'].astype(int)
    df_filtered['value'] = df_filtered['value'].astype(float)

    # Agrupar y calcular la media para duplicados
    df_filtered = df_filtered.groupby(['country', 'year']).agg({'value': 'mean'}).reset_index()

    return df_filtered
```

```python
def merge_dfs_safely(dfs):
    try:
        # Asegurarse de que no hay duplicados por pa√≠s (agrupar si hay varios valores)
        cleaned_dfs = []
        for i, df in enumerate(dfs):
            if df.duplicated('country').any():
                print(f"[INFO] Agrupando valores duplicados por 'country' en df {i}")
                df = df.groupby('country', as_index=False).mean(numeric_only=True)
            cleaned_dfs.append(df)

        # Unir todos con outer merge para no perder pa√≠ses
        df_final = reduce(lambda left, right: pd.merge(left, right, on='country', how='outer'), cleaned_dfs)

        # Mostrar tama√±o y shape final
        size_mb = df_final.memory_usage(deep=True).sum() / (1024 ** 2)
        print(f"\n‚úÖ Merge exitoso: {df_final.shape[0]} filas, {df_final.shape[1]} columnas")
        print(f"üì¶ Tama√±o aproximado: {size_mb:.2f} MB")
        return df_final

    except Exception as e:
        print("‚ùå Error al hacer el merge:", str(e))
        return None
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

def analizar_correlaciones(df, variable_objetivo):
    """
    df: DataFrame con columnas incluyendo 'country' y variables num√©ricas
    variable_objetivo: str, nombre de la columna que quieres analizar como target (no 'country')

    La funci√≥n:
    - Elimina 'country' y la variable objetivo de las features
    - Imputa valores faltantes con la media
    - Calcula y muestra las correlaciones absolutas de cada feature con la variable objetivo
    """
    if variable_objetivo not in df.columns:
        print(f"Error: '{variable_objetivo}' no est√° en el DataFrame.")
        return

    # Separar X e y
    X = df.drop(columns=['country', variable_objetivo])
    y = df[variable_objetivo]

    # Imputar valores faltantes
    imputer = SimpleImputer(strategy='mean')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

    # Calcular correlaci√≥n absoluta
    correlations = X_imputed.corrwith(y).abs().sort_values(ascending=False)

    return correlations

# Lista de variables objetivo
variables_objetivo = [
    'value_hale', 'value_life', 'value_under5', 'value_maternal',
    'value_malaria', 'value_tuberculosis', 'value_hiv',
    'value_suicide', 'value_nursing', 'value_doctors', 'value_airpollution', 'value_income'
]

# Diccionario para almacenar resultados
correlaciones_dict = {}
for var in variables_objetivo:
    correlaciones_dict[var] = analizar_correlaciones(df_final, var)

# Imprimir todos los resultados
for var, corr in correlaciones_dict.items():
    print(f"\nüìä Correlaciones absolutas con '{var}':\n{corr}")

```

## 5. M√©todos de An√°lisi

Para alcanzar el objetivo principal de este proyecto ‚Äîidentificar los 10 pa√≠ses con peor situaci√≥n de salud‚Äî hemos aplicado tres m√©todos de an√°lisis diferentes. Cada uno ofrece una perspectiva distinta sobre los datos, permiti√©ndonos contrastar resultados y validar su consistencia.

A continuaci√≥n se describe qu√© probamos y por qu√©:

‚úÖ **M√©todo 1 ‚Äî √çndice Compuesto con Signos Direccionales**

En este enfoque:
- Cada variable se estandariza (z-score) para que sean comparables en escala.
- Se construye un √≠ndice de salud combinando las variables con signos positivos o negativos seg√∫n su interpretaci√≥n:
  - Variables que indican peor salud si aumentan (mortalidad infantil, malaria, tuberculosis, HIV, suicidios, contaminaci√≥n) se suman con signo positivo.
  - Variables que indican mejor salud si aumentan (esperanza de vida, m√©dicos por habitante) se restan.

De este modo, el √≠ndice final refleja la contribuci√≥n neta de todos los factores, ponderados por su sentido l√≥gico.

üîπ Ventajas:
- Conserva la interpretaci√≥n directa de cada indicador.
- F√°cil de ajustar si se agregan m√°s variables.

üîπ Desventajas:
- No considera correlaciones entre variables.
- La elecci√≥n de signos requiere criterio experto y puede ser algo subjetiva.

‚úÖ **M√©todo 2 ‚Äî An√°lisis de Componentes Principales (PCA)**

Este m√©todo:
- Reduce la dimensionalidad del dataset.
- El primer componente principal capta la m√°xima variabilidad com√∫n entre los indicadores.
- Se utiliza este componente como un √≠ndice sint√©tico de salud.

üîπ Ventaja: Integra las correlaciones y relaciones lineales entre variables.

üîπ Desventaja: La interpretaci√≥n del componente puede no ser tan directa.

‚úÖ **M√©todo 3 ‚Äî Clusterizaci√≥n (K-Means)**

Aqu√≠:
- Se agrupan los pa√≠ses en un n√∫mero predefinido de clusters (por ejemplo, 3 grupos: salud buena, intermedia y mala).

- Identificamos el cluster con los valores medios m√°s desfavorables.

- Seleccionamos los pa√≠ses pertenecientes a ese grupo como los de peor situaci√≥n.

üîπ Ventaja: Permite segmentar sin imponer un ranking lineal.

üîπ Desventaja: El resultado depende de la elecci√≥n del n√∫mero de clusters y de la inicializaci√≥n.

### Com√∫n a los 3 M√©todos

##5.3. Clusterizaci√≥n K-Means

Aplicaci√≥n M√©todo Clusterizaci√≥n


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Elbow method
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 10), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()

# Silhouette score
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, labels)
    print(f"k={k}, silhouette score={score:.2f}")
```

**üéØ ¬øQu√© muestra este gr√°fico?**

- Eje X: n√∫mero de clusters (k), de 2 a 9.
- Eje Y: inercia, que es la suma de las distancias cuadradas de cada punto a su centroide.

**Menor inercia = clusters m√°s compactos.**

**üí° ¬øQu√© buscamos?**

Buscamos el punto donde la reducci√≥n de inercia deja de ser tan pronunciada (el ‚Äúcodo‚Äù):

- De k=2 a k=3, la inercia baja mucho.

- De k=3 a k=4, tambi√©n baja bastante.

- De k=4 a k=5, baja menos.

- De k=5 en adelante, la reducci√≥n es cada vez m√°s peque√±a.

**üîç ¬øD√≥nde est√° el ‚Äúcodo‚Äù?**

El ‚Äúcodo‚Äù es el punto despu√©s del cual la ganancia marginal se reduce claramente.
En el gr√°fico:
- ‚úÖ Entre k=3 y k=4 la pendiente ya empieza a suavizarse.
- ‚úÖ A partir de k=5 baja mucho menos.

Por tanto:
**üéØ k=3 o k=4 son los candidatos m√°s razonables.**

- k=3 un modelo m√°s simple.

- k=4 un poco m√°s de segmentaci√≥n.

Todos los indicadores fueron transformados previamente para que valores m√°s altos representaran peores condiciones de salud p√∫blica, independientemente de su naturaleza original.

```python
# Fem una c√≤pia per no modificar l'original
df_normalized = df_final_clean.copy()

# Normalitzaci√≥ min-max per a TOT el dataframe abans de filtrar
for col in cols:
    min_val = df_final_clean[col].min()
    max_val = df_final_clean[col].max()
    df_normalized[col] = (df_final_clean[col] - min_val) / (max_val - min_val)

# Ara df_normalized ja est√† normalitzat
print(df_normalized[cols].describe())
```

## 8.1. Simulaci√≥n de escenarios

‚úÖ OBJETIVO: Simular una mejora paulatina (temporal) en los indicadores clave para pa√≠ses seleccionados y calcular el nuevo PCA_health_score.

PASOS DE LA SIMULACI√ìN:

**1. Seleccionamos pa√≠ses a intervenir**
Usaremos estos pa√≠ses porque en el an√°lisis previo identificamos que tienen los peores indicadores de salud, y por tanto representan los casos m√°s urgentes o emblem√°ticos: ["Niger", "Chad", "Central African Republic", "Burkina Faso", "Lesotho", "South Africa"]

**2. Simulamos la mejora en los indicadores clave**

- Aumentamos value_doctors en un 100% (duplicar)
- Disminuimos value_under5, value_malaria, value_tuberculosis, value_hiv un 30%
- Aumentamos value_life un 20%
- Disminuimos value_suicide un 20% (Lesotho)
- Aumentamos gradualmente durante 5 a√±os

**3. Recalculamos el PCA_health_score**
Simularemos 5 a√±os y observaremos c√≥mo evoluciona el PCA_health_score.



```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Definimos los pa√≠ses a intervenir seg√∫n el an√°lisis previo
target_countries = ["Niger", "Chad", "Central African Republic", "Burkina Faso", "Lesotho", "South Africa"]

# Copia del dataframe original para no modificar datos base
df_sim = df_final_clean.copy()

# Diccionario para guardar resultados a√±o por a√±o
simulated_scores = {}

# Simulaci√≥n de 5 a√±os con mejoras graduales
for year in range(1, 6):
    df_temp = df_sim.copy()

    for country in target_countries:
        for col in cols:
            if col == 'value_doctors':
                # Incrementamos m√©dicos un 10% por a√±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 + 0.1 * year)
            elif col in ['value_under5', 'value_malaria', 'value_tuberculosis', 'value_hiv']:
                # Reducimos enfermedades y mortalidad infantil un 6% por a√±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 - 0.06 * year)
            elif col == 'value_hale':
                # Aumentamos esperanza de vida un 4% por a√±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 + 0.04 * year)
            elif col == 'value_suicide' and country == 'Lesotho':
                # Reducimos suicidio un 4% por a√±o solo en Lesotho
                df_temp.loc[df_temp["country"] == country, col] *= (1 - 0.04 * year)

    # Escalamos variables para PCA
    X_scaled = StandardScaler().fit_transform(df_temp[cols])

    # Aplicamos PCA para obtener el componente principal como score de salud
    pca = PCA(n_components=1)
    df_temp['PCA_health_score'] = pca.fit_transform(X_scaled)

    # Guardamos el score solo para pa√≠ses objetivo y a√±o actual
    simulated_scores[year] = df_temp[df_temp['country'].isin(target_countries)][['country', 'PCA_health_score']].set_index('country')

# Combinamos resultados a√±o a a√±o en un solo DataFrame
df_result = pd.concat(simulated_scores, axis=1)
df_result.columns = [f"Year_{i}" for i in range(1, 6)]
df_result = df_result.reset_index()

# Graficamos la evoluci√≥n del PCA_health_score por pa√≠s a lo largo de los 5 a√±os
plt.figure(figsize=(10, 6))
for country in target_countries:
    plt.plot(df_result.columns[1:], df_result[df_result['country'] == country].iloc[0, 1:], marker='o', label=country)

plt.title("Simulaci√≥n: Evoluci√≥n del PCA Health Score (5 a√±os)")
plt.xlabel("A√±o de intervenci√≥n")
plt.ylabel("PCA Health Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**üß† Interpretaci√≥n clave:**
- South Africa logra la mayor reducci√≥n absoluta del score (‚Äì1.40), lo que podr√≠a reflejar pol√≠ticas sanitarias eficaces.
- Chad y Burkina Faso parten con condiciones m√°s desfavorables, pero muestran grandes mejoras relativas.
- Todos los pa√≠ses convergen hacia valores m√°s bajos y similares, lo que sugiere una disminuci√≥n de desigualdades sanitarias si se mantiene esta tendencia.

## 8.2. Predicci√≥n de esperanza de vida

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Modelos a probar
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

# Variables predictoras y objetivo
X = df_final_clean[[
    'value_under5', 'value_malaria', 'value_tuberculosis',
    'value_hiv', 'value_suicide', 'value_doctors', 'value_airpollution','value_income'
]]
y = df_final_clean['value_hale']

# Separar datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Diccionario para guardar resultados
results = {}

# Lista de modelos a evaluar
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42, n_estimators=100),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42, n_estimators=100),
    'Support Vector Regressor': SVR(),
    'K-Neighbors Regressor': KNeighborsRegressor(n_neighbors=5)
}

# Entrenar y evaluar cada modelo
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results[name] = {'R2': r2, 'RMSE': rmse}
    print(f"{name} => R2: {r2:.3f}, RMSE: {rmse:.3f}")

# Mostrar resultados ordenados por R2 descendente
print("\nResumen de resultados:")
results_df = pd.DataFrame(results).T.sort_values(by='R2', ascending=False)
print(results_df)
```

Los resultados muestran que:

- Gradient Boosting es el mejor modelo, con un R¬≤ ‚âà 0.936 y un RMSE ‚âà 1.756, lo que indica una excelente capacidad predictiva y el menor error entre todos los modelos evaluados.

- Random Forest tambi√©n ofrece un desempe√±o muy s√≥lido, con un R¬≤ ‚âà 0.933 y un RMSE ‚âà 1.797, apenas ligeramente inferior al Gradient Boosting.

- Regresi√≥n Lineal obtiene un resultado aceptable, con un R¬≤ ‚âà 0.875 y un RMSE ‚âà 2.463, aunque su precisi√≥n es menor comparada con los modelos anteriores.

- K-Neighbors Regressor presenta un ajuste m√°s limitado, con un R¬≤ ‚âà 0.794 y un RMSE ‚âà 3.160, por lo que no resulta tan competitivo para este problema.

- Support Vector Regressor (SVR) es el que peor se adapta a los datos, con un R¬≤ ‚âà 0.556 y un RMSE ‚âà 4.643, mostrando una capacidad predictiva bastante limitada en este contexto.

```python
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV

# Modelo base
model = GradientBoostingRegressor(random_state=42)

# Validaci√≥n cruzada (5 folds) para obtener una evaluaci√≥n m√°s robusta del rendimiento.
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
print("CV R2 scores:", cv_scores)
print("Mean CV R2:", np.mean(cv_scores))

# B√∫squeda de hiperpar√°metros para intentar mejorar a√∫n m√°s.
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

#Interpretar el modelo: importancia de variables en Random Forest para entender qu√© factores afectan m√°s la esperanza de vida.
random_search = RandomizedSearchCV(
    model, param_distributions=param_dist,
    n_iter=20, cv=5, scoring='r2', random_state=42, n_jobs=-1, verbose=1
)
random_search.fit(X_train, y_train)

print("Best parameters:", random_search.best_params_)
print("Best CV R2:", random_search.best_score_)

# Modelo optimizado
best_model = random_search.best_estimator_

# Evaluar en test
y_pred = best_model.predict(X_test)
print("Test R2:", r2_score(y_test, y_pred))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# Importancia de variables
importances = best_model.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(8, 5))
plt.title("Importancia de variables - Random Forest")
plt.bar(range(len(importances)), importances[indices], color="b", align="center")
plt.xticks(range(len(importances)), features[indices], rotation=45)
plt.tight_layout()
plt.show()

# Gr√°fico predicciones vs valores reales
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Valor Real (Esperanza de vida)")
plt.ylabel("Predicci√≥n")
plt.title("Predicci√≥n vs Valor Real - Esperanza de Vida")
plt.grid(True)
plt.show()
```
