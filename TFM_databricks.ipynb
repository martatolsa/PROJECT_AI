## Life Expectancy & Healthy Life Expectancy (HALE)
**La esperanza de vida** proporciona una estimaciÃ³n de cuÃ¡ntos aÃ±os, en promedio, se espera que viva una poblaciÃ³n. Sin embargo, la **esperanza de vida saludable (HALE)** es un indicador mÃ¡s preciso del verdadero estado de salud de una poblaciÃ³n.

No se trata solo de la cantidad de aÃ±os que vive una persona en promedio, sino de cuÃ¡ntos de esos aÃ±os puede esperar vivir con buena salud, sin enfermedades o discapacidades graves que afecten su calidad de vida.

Este indicador es crucial porque muestra la diferencia entre simplemente vivir mÃ¡s aÃ±os y vivirlos en condiciones que permitan una vida plena y activa. Mientras que la esperanza de vida general mide la longevidad, **HALE tiene en cuenta factores como la carga de enfermedades, la discapacidad y el acceso a servicios de salud de calidad, proporcionando una visiÃ³n mÃ¡s completa del bienestar de una sociedad.**

## 3.1. ConstrucciÃ³n de funciones para limpiar y preprocesar

FunciÃ³n "preprocess_indicator" preprocesa un dataframe para un indicador especÃ­fico:
- Filtra segÃºn el indicador
- Renombra las columnas
- Elimina columnas irrelevantes
- Filtra la columna sex por "Both sexes"
- Asegura los tipos de datos de cada columna
- Agrupa los datos y calcula la media para las combinaciones duplicadas

```python

def preprocess_indicator(df, indicator_name):
    """
    Preprocesa un dataframe para un indicador especÃ­fico:
    - Filtra segÃºn el indicador
    - Renombra las columnas
    - Elimina columnas irrelevantes
    - Filtra la columna sex por "Both sexes"
    - Asegura los tipos de datos de cada columna
    - Agrupa los datos y calcula la media para las combinaciones duplicadas
    """
    # Filtrar datos segÃºn el indicador
    df_filtered = df[df['Indicator'] == indicator_name]

    # Asegurarse de que 'Dim1' estÃ© en las columnas, si no crearla
    if 'Dim1' not in df_filtered.columns:
        df_filtered['Dim1'] = 'Both sexes'

    columns_to_keep = [
    "Indicator", "ParentLocationCode", "ParentLocation", "SpatialDimValueCode",
    "Location", "Period", "Dim1" ,"FactValueNumeric", "Value"
    ]
    df_filtered = df_filtered[columns_to_keep]

    # Renombrar las columnas
    df_filtered = df_filtered.rename(columns={
        'Indicator': 'indicator',
        'ParentLocationCode': 'continent_code',
        'ParentLocation': 'continent',
        'SpatialDimValueCode': 'country_code',
        'Location': 'country',
        'Period': 'year',
        'Dim1': 'sex',
        'FactValueNumeric': 'value',
        'Value': 'value_interval'
    })

    # Filtrar por "Both sexes"
    df_filtered = df_filtered[df_filtered['sex'] == 'Both sexes']
    # Filtrar desde 1990 en adelante
    df_filtered = df_filtered[df_filtered['year'] >= 1990]

    # Eliminar columnas irrelevantes
    df_filtered = df_filtered[['continent','country', 'year', 'value']]

    # Asegurar tipos
    df_filtered['continent'] = df_filtered['continent'].astype(str)
    df_filtered['country'] = df_filtered['country'].astype(str)
    df_filtered['year'] = df_filtered['year'].astype(int)
    df_filtered['value'] = df_filtered['value'].astype(float)

    # Agrupar y calcular la media para duplicados
    df_filtered = df_filtered.groupby(['country', 'year']).agg({'value': 'mean'}).reset_index()

    return df_filtered
```

```python
def merge_dfs_safely(dfs):
    try:
        # Asegurarse de que no hay duplicados por paÃ­s (agrupar si hay varios valores)
        cleaned_dfs = []
        for i, df in enumerate(dfs):
            if df.duplicated('country').any():
                print(f"[INFO] Agrupando valores duplicados por 'country' en df {i}")
                df = df.groupby('country', as_index=False).mean(numeric_only=True)
            cleaned_dfs.append(df)

        # Unir todos con outer merge para no perder paÃ­ses
        df_final = reduce(lambda left, right: pd.merge(left, right, on='country', how='outer'), cleaned_dfs)

        # Mostrar tamaÃ±o y shape final
        size_mb = df_final.memory_usage(deep=True).sum() / (1024 ** 2)
        print(f"\nâœ… Merge exitoso: {df_final.shape[0]} filas, {df_final.shape[1]} columnas")
        print(f"ðŸ“¦ TamaÃ±o aproximado: {size_mb:.2f} MB")
        return df_final

    except Exception as e:
        print("âŒ Error al hacer el merge:", str(e))
        return None
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

def analizar_correlaciones(df, variable_objetivo):
    """
    df: DataFrame con columnas incluyendo 'country' y variables numÃ©ricas
    variable_objetivo: str, nombre de la columna que quieres analizar como target (no 'country')

    La funciÃ³n:
    - Elimina 'country' y la variable objetivo de las features
    - Imputa valores faltantes con la media
    - Calcula y muestra las correlaciones absolutas de cada feature con la variable objetivo
    """
    if variable_objetivo not in df.columns:
        print(f"Error: '{variable_objetivo}' no estÃ¡ en el DataFrame.")
        return

    # Separar X e y
    X = df.drop(columns=['country', variable_objetivo])
    y = df[variable_objetivo]

    # Imputar valores faltantes
    imputer = SimpleImputer(strategy='mean')
    X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

    # Calcular correlaciÃ³n absoluta
    correlations = X_imputed.corrwith(y).abs().sort_values(ascending=False)

    return correlations

# Lista de variables objetivo
variables_objetivo = [
    'value_hale', 'value_life', 'value_under5', 'value_maternal',
    'value_malaria', 'value_tuberculosis', 'value_hiv',
    'value_suicide', 'value_nursing', 'value_doctors', 'value_airpollution', 'value_income'
]

# Diccionario para almacenar resultados
correlaciones_dict = {}
for var in variables_objetivo:
    correlaciones_dict[var] = analizar_correlaciones(df_final, var)

# Imprimir todos los resultados
for var, corr in correlaciones_dict.items():
    print(f"\nðŸ“Š Correlaciones absolutas con '{var}':\n{corr}")

```

## 5. MÃ©todos de AnÃ¡lisi

Para alcanzar el objetivo principal de este proyecto â€”identificar los 10 paÃ­ses con peor situaciÃ³n de saludâ€” hemos aplicado tres mÃ©todos de anÃ¡lisis diferentes. Cada uno ofrece una perspectiva distinta sobre los datos, permitiÃ©ndonos contrastar resultados y validar su consistencia.

A continuaciÃ³n se describe quÃ© probamos y por quÃ©:

âœ… **MÃ©todo 1 â€” Ãndice Compuesto con Signos Direccionales**

En este enfoque:
- Cada variable se estandariza (z-score) para que sean comparables en escala.
- Se construye un Ã­ndice de salud combinando las variables con signos positivos o negativos segÃºn su interpretaciÃ³n:
  - Variables que indican peor salud si aumentan (mortalidad infantil, malaria, tuberculosis, HIV, suicidios, contaminaciÃ³n) se suman con signo positivo.
  - Variables que indican mejor salud si aumentan (esperanza de vida, mÃ©dicos por habitante) se restan.

De este modo, el Ã­ndice final refleja la contribuciÃ³n neta de todos los factores, ponderados por su sentido lÃ³gico.

ðŸ”¹ Ventajas:
- Conserva la interpretaciÃ³n directa de cada indicador.
- FÃ¡cil de ajustar si se agregan mÃ¡s variables.

ðŸ”¹ Desventajas:
- No considera correlaciones entre variables.
- La elecciÃ³n de signos requiere criterio experto y puede ser algo subjetiva.

âœ… **MÃ©todo 2 â€” AnÃ¡lisis de Componentes Principales (PCA)**

Este mÃ©todo:
- Reduce la dimensionalidad del dataset.
- El primer componente principal capta la mÃ¡xima variabilidad comÃºn entre los indicadores.
- Se utiliza este componente como un Ã­ndice sintÃ©tico de salud.

ðŸ”¹ Ventaja: Integra las correlaciones y relaciones lineales entre variables.

ðŸ”¹ Desventaja: La interpretaciÃ³n del componente puede no ser tan directa.

âœ… **MÃ©todo 3 â€” ClusterizaciÃ³n (K-Means)**

AquÃ­:
- Se agrupan los paÃ­ses en un nÃºmero predefinido de clusters (por ejemplo, 3 grupos: salud buena, intermedia y mala).

- Identificamos el cluster con los valores medios mÃ¡s desfavorables.

- Seleccionamos los paÃ­ses pertenecientes a ese grupo como los de peor situaciÃ³n.

ðŸ”¹ Ventaja: Permite segmentar sin imponer un ranking lineal.

ðŸ”¹ Desventaja: El resultado depende de la elecciÃ³n del nÃºmero de clusters y de la inicializaciÃ³n.

### ComÃºn a los 3 MÃ©todos

Limpieza de datos

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# SelecciÃ³n variables
cols = [
    'value_hale',
    'value_under5',
    'value_malaria',
    'value_tuberculosis',
    'value_hiv',
    'value_suicide',
    'value_doctors',
    'value_airpollution',
    'value_income']

# Descartar paÃ­ses con mÃ¡s de 5 NaN:
df_final_clean = df_final_clean[df_final_clean[cols].isnull().sum(axis=1) <= 5]

# Imputar los NaN restantes con la mediana de cada columna
df_final_clean[cols] = df_final_clean[cols].fillna(df_final_clean[cols].median())
```

Escalado o NormalizaciÃ³n

```python
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df_final_clean[cols])

# Convertir a DataFrame para mayor comodidad
scaled_df = pd.DataFrame(scaled_values, columns=cols, index=df_final_clean.index)
```

## 5.1 Ãndice Compuesto con Signos Direccionales

AplicaciÃ³n Ãndice Compuesto con Signos Direccionales

```python
df_final_clean['health_score'] = (
    - scaled_df['value_hale']               # Menor esperanza de vida = peor salud
    + scaled_df['value_under5']             # MÃ¡s mortalidad infantil = peor salud
    + scaled_df['value_malaria']
    + scaled_df['value_tuberculosis']
    + scaled_df['value_hiv']
    + scaled_df['value_suicide']
    - scaled_df['value_doctors']            # Menos mÃ©dicos = peor salud
    + scaled_df['value_airpollution']
    + scaled_df['value_income']             # Mayor ingreso = peor salud
)
```

IdentificaciÃ³n y SelecciÃ³n de los 10 PaÃ­ses con peor salud

```python
worst_health_score = df_final_clean.sort_values('health_score', ascending=False)

# Mostrar la tabla
print(worst_health_score[['country', 'health_score']])
```

## 5.2 PCA

AplicaciÃ³n MÃ©todo PCA

```python
# Aplicar PCA
pca = PCA(n_components=1)  # primer componente principal
principal_components = pca.fit_transform(scaled_values)

# AÃ±adir el score PCA al DataFrame original
df_final_clean['PCA_health_score'] = - principal_components[:, 0]

# Interpretar el componente principal:
print("Cargas del primer componente:")
for ind, comp in zip(cols, pca.components_[0]):
    print(f"{ind}: {comp:.3f}")

# Si las cargas son positivas y representan indicadores "malos" (ej: mortalidad), entonces mayor score PCA = peor salud.
# Si alguna carga estÃ¡ invertida, considera multiplicar por -1 para interpretaciÃ³n.

# Ordenar paÃ­ses segÃºn el score PCA (de mayor a menor)
worst_PCA_score = df_final_clean.sort_values('PCA_health_score', ascending=False)

# Mostrar la tabla
print(worst_PCA_score[['country', 'PCA_health_score']])
```

##5.3. ClusterizaciÃ³n K-Means

AplicaciÃ³n MÃ©todo ClusterizaciÃ³n


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Elbow method
inertia = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 10), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.show()

# Silhouette score
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, labels)
    print(f"k={k}, silhouette score={score:.2f}")
```

**ðŸŽ¯ Â¿QuÃ© muestra este grÃ¡fico?**

- Eje X: nÃºmero de clusters (k), de 2 a 9.
- Eje Y: inercia, que es la suma de las distancias cuadradas de cada punto a su centroide.

**Menor inercia = clusters mÃ¡s compactos.**

**ðŸ’¡ Â¿QuÃ© buscamos?**

Buscamos el punto donde la reducciÃ³n de inercia deja de ser tan pronunciada (el â€œcodoâ€):

- De k=2 a k=3, la inercia baja mucho.

- De k=3 a k=4, tambiÃ©n baja bastante.

- De k=4 a k=5, baja menos.

- De k=5 en adelante, la reducciÃ³n es cada vez mÃ¡s pequeÃ±a.

**ðŸ” Â¿DÃ³nde estÃ¡ el â€œcodoâ€?**

El â€œcodoâ€ es el punto despuÃ©s del cual la ganancia marginal se reduce claramente.
En el grÃ¡fico:
- âœ… Entre k=3 y k=4 la pendiente ya empieza a suavizarse.
- âœ… A partir de k=5 baja mucho menos.

Por tanto:
**ðŸŽ¯ k=3 o k=4 son los candidatos mÃ¡s razonables.**

- k=3 un modelo mÃ¡s simple.

- k=4 un poco mÃ¡s de segmentaciÃ³n.

Para mi objetivo (detectar los paÃ­ses con peor salud) y dado que la inercia sigue disminuyendo bastante hasta k=4:
**âœ… Elegimos k=4.**

```python
from sklearn.cluster import KMeans

# Puedes cambiar n_clusters segÃºn tu criterio
kmeans = KMeans(n_clusters=4, random_state=42)
df_final_clean['cluster'] = kmeans.fit_predict(scaled_df)
```

## 8.1. SimulaciÃ³n de escenarios

âœ… OBJETIVO: Simular una mejora paulatina (temporal) en los indicadores clave para paÃ­ses seleccionados y calcular el nuevo PCA_health_score.

PASOS DE LA SIMULACIÃ“N:

**1. Seleccionamos paÃ­ses a intervenir**
Usaremos estos paÃ­ses porque en el anÃ¡lisis previo identificamos que tienen los peores indicadores de salud, y por tanto representan los casos mÃ¡s urgentes o emblemÃ¡ticos: ["Niger", "Chad", "Central African Republic", "Burkina Faso", "Lesotho", "South Africa"]

**2. Simulamos la mejora en los indicadores clave**

- Aumentamos value_doctors en un 100% (duplicar)
- Disminuimos value_under5, value_malaria, value_tuberculosis, value_hiv un 30%
- Aumentamos value_life un 20%
- Disminuimos value_suicide un 20% (Lesotho)
- Aumentamos gradualmente durante 5 aÃ±os

**3. Recalculamos el PCA_health_score**
Simularemos 5 aÃ±os y observaremos cÃ³mo evoluciona el PCA_health_score.



```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Definimos los paÃ­ses a intervenir segÃºn el anÃ¡lisis previo
target_countries = ["Niger", "Chad", "Central African Republic", "Burkina Faso", "Lesotho", "South Africa"]

# Copia del dataframe original para no modificar datos base
df_sim = df_final_clean.copy()

# Diccionario para guardar resultados aÃ±o por aÃ±o
simulated_scores = {}

# SimulaciÃ³n de 5 aÃ±os con mejoras graduales
for year in range(1, 6):
    df_temp = df_sim.copy()

    for country in target_countries:
        for col in cols:
            if col == 'value_doctors':
                # Incrementamos mÃ©dicos un 10% por aÃ±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 + 0.1 * year)
            elif col in ['value_under5', 'value_malaria', 'value_tuberculosis', 'value_hiv']:
                # Reducimos enfermedades y mortalidad infantil un 6% por aÃ±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 - 0.06 * year)
            elif col == 'value_hale':
                # Aumentamos esperanza de vida un 4% por aÃ±o
                df_temp.loc[df_temp["country"] == country, col] *= (1 + 0.04 * year)
            elif col == 'value_suicide' and country == 'Lesotho':
                # Reducimos suicidio un 4% por aÃ±o solo en Lesotho
                df_temp.loc[df_temp["country"] == country, col] *= (1 - 0.04 * year)

    # Escalamos variables para PCA
    X_scaled = StandardScaler().fit_transform(df_temp[cols])

    # Aplicamos PCA para obtener el componente principal como score de salud
    pca = PCA(n_components=1)
    df_temp['PCA_health_score'] = pca.fit_transform(X_scaled)

    # Guardamos el score solo para paÃ­ses objetivo y aÃ±o actual
    simulated_scores[year] = df_temp[df_temp['country'].isin(target_countries)][['country', 'PCA_health_score']].set_index('country')

# Combinamos resultados aÃ±o a aÃ±o en un solo DataFrame
df_result = pd.concat(simulated_scores, axis=1)
df_result.columns = [f"Year_{i}" for i in range(1, 6)]
df_result = df_result.reset_index()

# Graficamos la evoluciÃ³n del PCA_health_score por paÃ­s a lo largo de los 5 aÃ±os
plt.figure(figsize=(10, 6))
for country in target_countries:
    plt.plot(df_result.columns[1:], df_result[df_result['country'] == country].iloc[0, 1:], marker='o', label=country)

plt.title("SimulaciÃ³n: EvoluciÃ³n del PCA Health Score (5 aÃ±os)")
plt.xlabel("AÃ±o de intervenciÃ³n")
plt.ylabel("PCA Health Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**ðŸ§  InterpretaciÃ³n clave:**
- South Africa logra la mayor reducciÃ³n absoluta del score (â€“1.40), lo que podrÃ­a reflejar polÃ­ticas sanitarias eficaces.
- Chad y Burkina Faso parten con condiciones mÃ¡s desfavorables, pero muestran grandes mejoras relativas.
- Todos los paÃ­ses convergen hacia valores mÃ¡s bajos y similares, lo que sugiere una disminuciÃ³n de desigualdades sanitarias si se mantiene esta tendencia.

## 8.2. PredicciÃ³n de esperanza de vida

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Modelos a probar
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

# Variables predictoras y objetivo
X = df_final_clean[[
    'value_under5', 'value_malaria', 'value_tuberculosis',
    'value_hiv', 'value_suicide', 'value_doctors', 'value_airpollution','value_income'
]]
y = df_final_clean['value_hale']

# Separar datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Diccionario para guardar resultados
results = {}

# Lista de modelos a evaluar
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42, n_estimators=100),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42, n_estimators=100),
    'Support Vector Regressor': SVR(),
    'K-Neighbors Regressor': KNeighborsRegressor(n_neighbors=5)
}

# Entrenar y evaluar cada modelo
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results[name] = {'R2': r2, 'RMSE': rmse}
    print(f"{name} => R2: {r2:.3f}, RMSE: {rmse:.3f}")

# Mostrar resultados ordenados por R2 descendente
print("\nResumen de resultados:")
results_df = pd.DataFrame(results).T.sort_values(by='R2', ascending=False)
print(results_df)
```

Los resultados muestran que:

- Gradient Boosting es el mejor modelo, con un RÂ² â‰ˆ 0.936 y un RMSE â‰ˆ 1.756, lo que indica una excelente capacidad predictiva y el menor error entre todos los modelos evaluados.

- Random Forest tambiÃ©n ofrece un desempeÃ±o muy sÃ³lido, con un RÂ² â‰ˆ 0.933 y un RMSE â‰ˆ 1.797, apenas ligeramente inferior al Gradient Boosting.

- RegresiÃ³n Lineal obtiene un resultado aceptable, con un RÂ² â‰ˆ 0.875 y un RMSE â‰ˆ 2.463, aunque su precisiÃ³n es menor comparada con los modelos anteriores.

- K-Neighbors Regressor presenta un ajuste mÃ¡s limitado, con un RÂ² â‰ˆ 0.794 y un RMSE â‰ˆ 3.160, por lo que no resulta tan competitivo para este problema.

- Support Vector Regressor (SVR) es el que peor se adapta a los datos, con un RÂ² â‰ˆ 0.556 y un RMSE â‰ˆ 4.643, mostrando una capacidad predictiva bastante limitada en este contexto.

```python
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV

# Modelo base
model = GradientBoostingRegressor(random_state=42)

# ValidaciÃ³n cruzada (5 folds) para obtener una evaluaciÃ³n mÃ¡s robusta del rendimiento.
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
print("CV R2 scores:", cv_scores)
print("Mean CV R2:", np.mean(cv_scores))

# BÃºsqueda de hiperparÃ¡metros para intentar mejorar aÃºn mÃ¡s.
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

#Interpretar el modelo: importancia de variables en Random Forest para entender quÃ© factores afectan mÃ¡s la esperanza de vida.
random_search = RandomizedSearchCV(
    model, param_distributions=param_dist,
    n_iter=20, cv=5, scoring='r2', random_state=42, n_jobs=-1, verbose=1
)
random_search.fit(X_train, y_train)

print("Best parameters:", random_search.best_params_)
print("Best CV R2:", random_search.best_score_)

# Modelo optimizado
best_model = random_search.best_estimator_

# Evaluar en test
y_pred = best_model.predict(X_test)
print("Test R2:", r2_score(y_test, y_pred))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# Importancia de variables
importances = best_model.feature_importances_
features = X.columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(8, 5))
plt.title("Importancia de variables - Random Forest")
plt.bar(range(len(importances)), importances[indices], color="b", align="center")
plt.xticks(range(len(importances)), features[indices], rotation=45)
plt.tight_layout()
plt.show()

# GrÃ¡fico predicciones vs valores reales
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Valor Real (Esperanza de vida)")
plt.ylabel("PredicciÃ³n")
plt.title("PredicciÃ³n vs Valor Real - Esperanza de Vida")
plt.grid(True)
plt.show()
```
